```python
%%writefile simd_example.c
#include <stdio.h>
#include <omp.h>

#define N 16

int main() {
  int i;
  float a[N], b[N], c[N];

  // Intialize arrays
  for (i = 0; i < N; i++) {
    a[i] = i * 1.0;
    b[i] = i * 2.0;
}

// SIMD Vectorization
#pragma omp simd
for (i = 0; i < N; i++) {
  c[i] = a[i] + b[i];
}

// Print result
for (i = 0; i < N; i++) {
  printf("c[%d] = %.2f\n", i, c[i]);
}
return 0;
}
```

```python
%%writefile vector_add.c
#include <stdio.h>
#include<omp.h>

#define N 1024

void Vector_add(float* a, float* b, float* c)
{
    int i;
    for (i = 0; i < N; i++) {
        c[i] = a[i] + b[i];
    }
  }

int main() {
      float a[N], b[N], c[N];

      for (int i = 0; i < N; i++) {
          a[i] = (float)i * 1.5f;
          b[i] = (float)i * 2.3f;
          c[i] = 0.0f;
      }

      Vector_add(a, b, c);

      for (int i = 0; i < N; i++) {
          if (c[i] != a[i] + b[i]) {
              printf("Error at index %d!\n", i);
              return 1;
          }
      }

      printf("Vector addition successful with openMP SIMD.\n");

      return 0;
  }
```
